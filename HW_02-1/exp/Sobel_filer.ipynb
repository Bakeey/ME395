```python
from PIL import Image
from numpy import *
from pylab import *
from scipy.ndimage import filters

im = array(Image.open('Lecture_1_image.jpg').convert('L'))

# Sobel derivative filters
imx = zeros(im.shape)
filters.sobel(im,1,imx)

imy = zeros(im.shape)
filters.sobel(im,0,imy)

magnitude = sqrt(imx**2+imy**2)

figure(figsize=(20,20))
gray()

subplot(1,2,1)
title('Oiginal')
axis('off')
imshow(im)

#subplot(1,4,2)
#title('imx')
#axis('off')
#imshow(imx)

#subplot(1,4,3)
#title('imy')
#axis('off')
#imshow(imy)

subplot(1,2,2)
title('magnitude')
axis('off')
imshow(magnitude)

savefig('sobel.png')
show()
```
